```python
#import gcloud aiplatform 
from google.cloud import aiplatform

#import gcloud storage
from google.cloud import storage

#import langchain 
```

```python
#assign project and location
aiplatform.init(project='automatedanomaly', location='us-central1')
```

```python
#initialize client 
client = storage.Client()
```

```python
#identify and load data from bucket 
#bucketname = baxterbucket1
bucketname = 'baxterbucket1'
blob_path = 'testing.txt'
```

```python
#initialize bucket and blob
bucket = client.bucket(bucketname)
blob = bucket.blob(blob_path)
```

```python
#read directly into memory and print
content = blob.download_as_text()
print(content) 
```
